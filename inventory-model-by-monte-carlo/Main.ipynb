# The Mian Problem 
we need to predict the total cost of one inventory depending on old data.

## Demand DataFrame

### Create the DataFrame from Existed Data

```python
#import pandas to start working with
import pandas as pd

# take data to create dataframe with
demand_df_data = {
    "demand": [0, 1, 2, 3, 4, 5],
    "frequency": [15, 30, 60, 120, 45, 30]
}

# create data frame 
demand_df = pd.DataFrame(demand_df_data)


demand_df 
```

### Adding Probability Column

```python
# now let's add probability column

# first calculate the summition of ferquency column to find something to devide frequency into to get the probabitliy
total_days = sum(demand_df["frequency"])

demand_df["probability"] = demand_df["frequency"] / total_days
demand_df
```

### Adding Commutative Probability to the DataFrame

```python
# now the hardest is to get commutative probability

# to get the sum of all before rows
prob = 0

# this operarion requires working with rows
for row in range(demand_df.shape[0]):
    
    # like sum of the whole previous row's probability
    prob = demand_df.loc[row, "probability"] + prob
    
    # assign them to the new row
    demand_df.loc[row, "commutative probability"] = prob
    
demand_df

# print(demand_df.shape)
```

### Creating Ranges Depending on Commutative Probability

> the problem is i can't put a list in one cell as i do on paper, so i separated them into two columns
**interval start** and **interval end**

```python
# it's time to make ranges

# the first row always starts with 1
start = 1

# loop through all rows -- this operation should happen on rows 
for row in range(demand_df.shape[0]):
    # the start column is the same start
    # using .loc is faster and safer so i used them the old version is commented
    
    # demand_df["interval start"][row] = start
    demand_df.loc[row, "interval start"] = start
    
    # the end container value -- commutative probability * 100
    end = demand_df["commutative probability"][row] * 100
    
    # the end is the end container value -- two steps for readability
    # using .loc is faster and safer so i used them the old version is commented
    
    # demand_df["interval end"][row] = end
    demand_df.loc[row, "interval end"] = end
    
    # update the start to become one step after the end
    start = end + 1
    
demand_df
```

## the DataFrame is now Ready to Work with

### the needed columns are
1. demand
2. interval start
3. interval end


```python
needed_demand_df = pd.concat([demand_df["demand"], demand_df["interval start"], demand_df["interval end"]], axis=1)
needed_demand_df
```
